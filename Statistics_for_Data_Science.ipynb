```python
from collections import Counter
# Raw data
data = [91, 79, 66, 98, 127, 139, 154, 147, 192, 88, 97, 92, 87, 142, 127, 184, 145, 162, 95, 89, 86, 98, 145, 129, 149, 158, 241]
# Intervals for frequency distribution
intervals = [(66, 87), (88, 109), (220, 241)]
# Frequency distribution calculation
frequency_distribution = Counter()
for num in data:
    for interval in intervals:
        if interval[0] <= num <= interval[1]:
            frequency_distribution[interval] += 1
            break
# Modal value
modal_value_interval = max(frequency_distribution, key=frequency_distribution.get)
modal_value = sum(modal_value_interval) / 2
# Mean calculation
mean = sum(data) / len(data)
# Output
print("Frequency distribution:")
for interval, frequency in sorted(frequency_distribution.items(), key=lambda x: x[0]):
    print(f"{interval[0]}-{interval[1]}: {frequency}")
print("\nModal value:", modal_value)
print("Mean:", mean)
# Comparison
if modal_value == modal_value_interval[0] and modal_value == modal_value_interval[1]:
    print("\nModal value represents the most frequently occurring range.")
elif modal_value < mean:
    print("\nMean is higher than modal value.")
else:
    print("\nModal value is higher than mean.")
```

```python
#Part a,b,c
import numpy as np
# Test scores for the day and night classes
day_scores = np.array([99, 56, 78, 55.5, 32, 90, 80, 81, 56, 59, 45, 77, 84.5, 84, 70, 72, 68, 32, 79, 90])
night_scores = np.array([98, 78, 68, 83, 81, 89, 88, 76, 65, 45, 98, 90, 80, 84.5, 85, 79, 78, 98, 90, 79, 81, 25.5])
# Calculate statistics for both classes
def calculate_statistics(scores):
    return np.min(scores), np.max(scores), np.median(scores), np.percentile(scores, 25), np.percentile(scores, 75)
# Calculate percentages for both classes
def calculate_percentages(scores, min_score, first_quartile, median, third_quartile, max_score):
    return [np.count_nonzero((scores >= min_score) & (scores <= first_quartile)) / len(scores) * 100,
            np.count_nonzero((scores >= first_quartile) & (scores <= median)) / len(scores) * 100,
            np.count_nonzero((scores >= median) & (scores <= third_quartile)) / len(scores) * 100,
            np.count_nonzero((scores >= third_quartile) & (scores <= max_score)) / len(scores) * 100,
            np.count_nonzero((scores >= first_quartile) & (scores <= max_score)) / len(scores) * 100]
# Calculate statistics
day_min, day_max, day_median, day_first_quartile, day_third_quartile = calculate_statistics(day_scores)
night_min, night_max, night_median, night_first_quartile, night_third_quartile = calculate_statistics(night_scores)
# Calculate percentages
day_percentages = calculate_percentages(day_scores, day_min, day_first_quartile, day_median, day_third_quartile, day_max)
night_percentages = calculate_percentages(night_scores, night_min, night_first_quartile, night_median, night_third_quartile, night_max)
# Printing results
print("Day Class Statistics:")
print("Smallest value:", day_min)
print("Largest value:", day_max)
print("Median:", day_median)
print("First Quartile:", day_first_quartile)
print("Third Quartile:", day_third_quartile)
print("\nNight Class Statistics:")
print("Smallest value:", night_min)
print("Largest value:", night_max)
print("Median:", night_median)
print("First Quartile:", night_first_quartile)
print("Third Quartile:", night_third_quartile)
print("\nPercentages for Day Class:")
print("Between smallest value and first quartile:", day_percentages[0])
print("Between first quartile and median:", day_percentages[1])
print("Between median and third quartile:", day_percentages[2])
print("Between third quartile and largest value:", day_percentages[3])
print("Between first quartile and largest value:", day_percentages[4])
print("\nPercentages for Night Class:")
print("Between smallest value and first quartile:", night_percentages[0])
print("Between first quartile and median:", night_percentages[1])
print("Between median and third quartile:", night_percentages[2])
print("Between third quartile and largest value:", night_percentages[3])
print("Between first quartile and largest value:", night_percentages[4])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 10000  # Size of the population
sample_size = 30         # Size of each sample
num_samples = 1000       # Number of samples to generate

# Generate normal population
population_mean = 100
population_std = 10
population = np.random.normal(population_mean, population_std, population_size)

# Generate sample means
sample_means = [np.mean(np.random.choice(population, size=sample_size)) for _ in range(num_samples)]

# Plot histogram of sample means
plt.figure(figsize=(8, 6))
plt.hist(sample_means, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7)

# Fit a normal distribution to the sample means
mu = np.mean(sample_means)
sigma = np.std(sample_means)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(- (x - mu)**2 / (2 * sigma**2)), color='red', linestyle='--')

plt.title('Sampling Distribution of Sample Means (CLT Verification)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend(['Normal Distribution', 'Sample Means'])
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 1000     # Size of the population
true_proportion = 0.395    # True population proportion for smoking
num_samples = 10000         # Number of samples to generate
sample_size = 50           # Size of each sample

# Generate population data
population = np.random.choice([0, 1], size=population_size, p=[1-true_proportion, true_proportion])

# Generate sample proportions
sample_proportions = [np.mean(np.random.choice(population, size=sample_size)) for _ in range(num_samples)]

# Plot histogram of sample proportions
plt.figure(figsize=(8, 6))
plt.hist(sample_proportions, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7)

# Calculate the mean and standard deviation of the sample proportions
sample_mean = np.mean(sample_proportions)
sample_std = np.std(sample_proportions)

# Fit a normal distribution to the sample proportions
x = np.linspace(sample_mean - 3*sample_std, sample_mean + 3*sample_std, 100)
plt.plot(x, 1/(sample_std * np.sqrt(2 * np.pi)) * np.exp(- (x - sample_mean)**2 / (2 * sample_std**2)), color='red', linestyle='--')

plt.title('Sampling Distribution of Sample Proportions (CLT Verification)')
plt.xlabel('Sample Proportion')
plt.ylabel('Density')
plt.legend(['Normal Distribution', 'Sample Proportions'])
plt.grid(True)
plt.show()
```

```python
def chi2_ind_att(df,alpha):
  row_total = sum(df.values.T)
  column_total = sum(df.values)
  total = sum(row_total)
  obs_freq = []
  exp_freq = []
  chi2_stats = 0
  m = len(row_total)
  n = len(column_total)
  for i in range(m):
    for j in range(n):
      exp = round(row_total[i]*column_total[j]/total)
      exp_freq.append(exp)
      obs = df.values[i][j]
      obs_freq.append(obs)
      chi2_stats = chi2_stats + (obs -exp)**2/exp
  print('Observed Frequency =',obs_freq)
  print("Total Observed Frequency = " + str(sum(obs_freq)))
  print('Expected Frequency =',exp_freq)
  print("Total Expected Frequency = " + str(sum(exp_freq)))
  dof = (m-1)*(n-1)
  p_value = chi2.sf(chi2_stats,dof)
  print('chisquare_statistics =',chi2_stats, 'and p_value =',p_value)
  if p_value > alpha:
    print("Failed to reject the null hypothesis for level of significance = " + str(alpha))
  else:
    print("Null hypothesis is rejected for level of significance = " + str(alpha))
```

```python
print("Null Hypothesis is that gender and survival of passengers are independent of others.")
print("Alternate Hypothesis is that gender and survival of passengers are related to each other.")
alpha = 0.05
chi2_ind_att(df, alpha)
```

```python
path = "/content/drive/MyDrive/bulb_life.csv"
data = pd.read_csv(path)
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(data=data,x='life',bins=8)
plt.show()
```

```python
from scipy.stats import norm
mean = np.mean(data['life'])
std = np.std(data['life'])
bins = 8
interval = []
for i in range(1,9):
  val = norm.ppf(i/bins,mean,std)
  interval.append(val)
interval
```

```python
interval.insert(0, -np.inf)
interval
```

```python
df = pd.DataFrame({'lower_limit' : interval[:-1],'upper_limit' : interval[1:]})
df
```

```python
life_values = list(sorted(data['life']))
df['obs_freq'] = df.apply(lambda x:sum([i>x['lower_limit'] and i<=x['upper_limit'] for i in life_values]),axis = 1)
df['exp_freq'] = 5
df
```

```python
print("Null Hypothesis is that sample data follows the normal distribution.")
print("Alternate Hypothesis is that sample data  does not follow normal distribution.")
alpha = 0.05
chi2_stats , p_value = scipy.stats.chisquare(df['obs_freq'],df['exp_freq'])
print('Chisquare_statistics = ',chi2_stats,'and p_value = ',p_value)
if p_value > alpha:
  print("Failed to reject the null hypothesis for the level of significance " + str(alpha))
else:
  print("Null hypothesis is rejected for the level of significance " + str(alpha))
```
