```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
# Fetch stock data
def get_stock_data(symbol, start_date, end_date):
    df = yf.download(symbol, start=start_date, end=end_date)
    return df[['Close']]
```

```python
# Plot closing price
def plot_data(df):
    plt.figure(figsize=(10,5))
    plt.plot(df.index, df['Close'], label='Closing Price')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```

```python
# Split train-test (80-20)
def split_data(df):
    train_size = int(len(df) * 0.8)
    return df[:train_size], df[train_size:]
```

```python
# Scale data
def scale_data(train, test):
    scaler = MinMaxScaler(feature_range=(0,1))
    train_scaled = scaler.fit_transform(train)
    test_scaled = scaler.transform(test)
    return scaler, train_scaled, test_scaled
```

```python
# Prepare sequences
def create_sequences(data, seq_length=25):
    sequences, labels = [], []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i+seq_length])
        labels.append(data[i+seq_length])
    return np.array(sequences), np.array(labels)
```

```python
# Build LSTM model
def build_lstm_model(input_shape):
    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=input_shape),
        LSTM(50, return_sequences=False),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# Plot predictions
def plot_predictions(train, test_actual, test_pred):
    plt.figure(figsize=(10,5))
    plt.plot(train, label='Training Data')
    plt.plot(range(len(train), len(train) + len(test_actual)), test_actual, label='Actual Test Data', color='orange')
    plt.plot(range(len(train), len(train) + len(test_pred)), test_pred, label='Predictions', color='green')
    plt.legend()
    plt.show()
```

```python
# Evaluate model
def evaluate_model(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    print(f'MAE: {mae}, RMSE: {rmse}')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Main Execution
symbol = 'AAPL'
start_date, end_date = '2023-01-01', '2024-03-01'
df = get_stock_data(symbol, start_date, end_date)
plot_data(df)
train_data, test_data = split_data(df)
scaler, train_scaled, test_scaled = scale_data(train_data, test_data)
X_train, y_train = create_sequences(train_scaled)
X_test, y_test = create_sequences(test_scaled)
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
model = build_lstm_model((X_train.shape[1], 1))
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)
y_test_pred = model.predict(X_test)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1,1))
test_pred_inv = scaler.inverse_transform(y_test_pred)
plot_predictions(train_data.values[25:], y_test_actual, test_pred_inv)
evaluate_model(y_test_actual, test_pred_inv)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating the data dictionary
data = {
    'Fund': ['A', 'B', 'C', 'D', 'E'],
    'Monthly Returns': [
        [0.03, 0.02, 0.05, 0.04, 0.03, 0.01, -0.02, 0.04, 0.02, 0.03, 0.02, 0.04],
        [0.02, 0.01, 0.04, 0.03, 0.02, 0.01, -0.01, 0.03, 0.01, 0.02, 0.01, 0.03],
        [0.05, 0.04, 0.06, 0.05, 0.04, 0.03, 0.02, 0.06, 0.04, 0.05, 0.03, 0.04],
        [0.03, 0.04, 0.05, 0.03, 0.02, 0.01, 0.00, 0.03, 0.01, 0.02, 0.04, 0.05],
        [0.02, 0.03, 0.04, 0.03, 0.01, 0.02, -0.01, 0.03, 0.02, 0.01, 0.02, 0.03]
    ],
    'Beta': [1.2, 0.8, 1.0, 1.1, 1.3]
}
```

```python
data
```

```python
# Converting to DataFrame
returns_df = pd.DataFrame(data)
```

```python
returns_df
```

```python
# Risk-free rate (annualized)
risk_free_rate = 0.02
```

```python
# Function to calculate annualized return
def annualized_return(monthly_returns):
    return np.prod([1 + r for r in monthly_returns]) ** (12 / len(monthly_returns)) - 1
```

```python
def calculate_sharpe_ratio(monthly_returns, risk_free_rate):
    ann_return = annualized_return(monthly_returns)
    ann_volatility = np.std(monthly_returns) * np.sqrt(12)
    sharpe_ratio = np.mean(ann_return - risk_free_rate) / ann_volatility
    return sharpe_ratio
```

```python
def calculate_treynor_ratio(monthly_returns, beta, risk_free_rate):
    ann_return = annualized_return(monthly_returns)
    treynor_ratio = (ann_return - risk_free_rate) / beta
    return treynor_ratio
```

```python
# Calculating metrics for each fund
metrics = []
for i in range(len(returns_df)):
    fund = returns_df.loc[i, 'Fund']
    monthly_returns = returns_df.loc[i, 'Monthly Returns']
    beta = returns_df.loc[i, 'Beta']
    ann_return = annualized_return(monthly_returns)
    sharpe = calculate_sharpe_ratio(monthly_returns, risk_free_rate)
    treynor = calculate_treynor_ratio(monthly_returns, beta, risk_free_rate)
    metrics.append({'Fund': fund, 'Annualized Return': ann_return, 'Sharpe Ratio': sharpe, 'Treynor Ratio': treynor})
```

```python
# Converting metrics to DataFrame
metrics_df = pd.DataFrame(metrics).set_index('Fund')
metrics_df
```

```python
# Visualization
metrics_df.plot(kind='bar', figsize=(10, 6))
plt.title("Comparison of Mutual Funds Based on Key Metrics")
plt.ylabel("Value")
plt.xlabel("Fund")
plt.legend(loc='best')
plt.grid(axis='y')
plt.show()
```

```python
# Identifying the best fund based on Sharpe Ratio
best_fund = metrics_df['Sharpe Ratio'].idxmax()
print(f"The best fund based on Sharpe Ratio is: Fund {best_fund}")
```

```python
def analyze_financials(ticker):
  # Fetch Data From Yahoo Finance
  stock = yf.Ticker(ticker)
  # Get Balance Sheet, CashFlow and Profit and Loss Statements
  balance_sheet = stock.balance_sheet
  cash_flow = stock.cashflow
  income_statement = stock.financials
  # Display the data
  print(f"Financial data for {ticker}:")
  print("\n--- Balance Sheet ---")
  print(balance_sheet)
  print("\n--- Cash Flow Statement ---")
  print(cash_flow)
  print("\n--- Income Statement ---")
  print(income_statement)
```

```python
def analyze_balance_sheet(balance_sheet):
    # Check if the input is a pandas DataFrame
    if not isinstance(balance_sheet, pd.DataFrame):
        print("Error: Input must be a pandas DataFrame.")
        return
    try:
        # Iterate through dates in 2024
        for date in balance_sheet.columns:
          if date.year == 2024:  # Check if the year is 2024
              try:
                  total_assets = balance_sheet[date]['Total Assets']
                  total_liabilities = balance_sheet[date]['Total Liabilities Net Minority Interest']
                  shareholder_equity = total_assets - total_liabilities
                  print(f"Balance Sheet Analysis for {date}:")
                  print(f"Total Assets: {total_assets}")
                  print(f"Total Liabilities: {total_liabilities}")
                  print(f"Shareholder's Equity: {shareholder_equity}")
                  print("---")
              except KeyError as e:
                  print(f"Error: Key not found in balance sheet for {date} - {e}")
    except AttributeError:
        print("Error: The balance_sheet DataFrame may not have columns with date information")
```

```python
def analyze_cash_flow(cash_flow):
    try:
        operating_cash_flow = cash_flow.loc['Operating Cash Flow', '2024-09-30']
    except KeyError:
        print("Error: 'Operating Cash Flow' for 2024 not found in cash flow statement.")
        return
    try:
        capital_expenditures = cash_flow.loc['Capital Expenditure', '2024-09-30']
    except KeyError:
        try:
            capital_expenditures = cash_flow.loc['Investments', '2024-09-30']
            print("'Capital Expenditures' not found for 2024, using 'Investments' instead.")
        except KeyError:
            capital_expenditures = 0
            print("'Capital Expenditures' and 'Investments' not found for 2024, defaulting to 0.")
    free_cash_flow = operating_cash_flow + capital_expenditures
    print("\n--- Cash Flow Analysis for 2024 ---")
    print(f"Operating Cash Flow: {operating_cash_flow}")
    print(f"Capital Expenditures: {capital_expenditures}")
    print(f"Free Cash Flow: {free_cash_flow}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
# Given data
maturities = np.array([0.25, 0.5, 1, 2, 3, 5, 7, 10, 20, 30])
yields = np.array([3.8, 3.9, 4.0, 4.2, 4.3, 4.5, 4.7, 4.8, 5.0, 5.1])
```

```python
# Function to smooth the curve
def smooth_curve(x, y, num_points=300):
    x_smooth = np.linspace(x.min(), x.max(), num_points)
    spline = make_interp_spline(x, y, k=3)
    y_smooth = spline(x_smooth)
    return x_smooth, y_smooth
```

```python
# Generate smoothed data
x_smooth, y_smooth = smooth_curve(maturities, yields)
```

```python
# Create a DataFrame
data = {'x_smooth': x_smooth, 'y_smooth': y_smooth}
df_smooth = pd.DataFrame(data)
```

```python
# Display the table
df_smooth
```

```python
# Plot the yield curve
plt.figure(figsize=(10, 6))
plt.plot(x_smooth, y_smooth, label='Smoothed Yield Curve', color='blue')
plt.scatter(maturities, yields, color='red', label='Data Points')
# Labels, title, and legend
plt.xlabel("Maturities (Years)")
plt.ylabel("Yields (%)")
plt.title("Yield Curve Analysis")
plt.legend()
plt.grid()
# Display the plot
plt.show()
```

```python
# Yield curve analysis
if yields[-1] > yields[0]:
    print("The yield curve is upward-sloping, indicating economic growth expectations.")
elif yields[-1] < yields[0]:
    print("The yield curve is inverted, which might signal a potential recession.")
else:
    print("The yield curve is flat, suggesting economic uncertainty.")
```

```python
# c. Import datetime tools
from datetime import datetime, timedelta
```

```python
# Step 1: Download USD to INR Data
def download_usd_inr_data(start_date='2015-01-01', end_date=None):
    if end_date is None:
        end_date = datetime.today().strftime('%Y-%m-%d')
    data = yf.download('USDINR=X', start=start_date, end=end_date)
    df = data[['Close']].copy()
    df.rename(columns={'Close': 'USDINR'}, inplace=True)
    return df
```

```python
# Step 2: Train ARIMA Model
def train_arima_model(data, order=(1, 2, 1)):
    model = ARIMA(data, order=order)
    model_fit = model.fit()
    return model_fit
```

```python
# Step 3: Forecast Future Values
def forecast_future(model_fit, steps=30):
    forecast = model_fit.forecast(steps=steps)
    return forecast
```

```python
# Step 4: Plot Historical and Forecasted Data
def plot_forecast(data, forecast):
    plt.figure(figsize=(14, 6))
    plt.plot(data.index, data['USDINR'], label='Historical Data')
    last_date = data.index[-1]
    future_dates = pd.date_range(start=last_date + timedelta(days=1), periods=len(forecast), freq='B')
    plt.plot(future_dates, forecast, color='red', label='Forecasted Data')
    plt.title('USD to INR Exchange Rate Prediction using ARIMA')
    plt.xlabel('Date')
    plt.ylabel('Exchange Rate (USD to INR)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Define Portfolio Parameters
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN']
weights = [0.3, 0.3, 0.2, 0.2]
start_date = '2018-01-01'
end_date = '2023-12-31'
num_simulations = 10000
trading_days = 252
```

```python
# Fetch Historical Stock Data
data = yf.download(tickers, start=start_date, end=end_date)
```

```python
# Extract appropriate price data
if 'Adj Close' in data.columns:
    price_data = data['Adj Close']
elif 'Close' in data.columns:
    price_data = data['Close']
else:
    raise KeyError("Neither 'Adj Close' nor 'Close' found in the downloaded data.")
```

```python
# Calculate Daily Returns
daily_returns = price_data.pct_change().dropna()
```

```python
# Show the first few rows of daily returns
daily_returns.head()
```

```python
# Calculate the mean of daily returns for each stock
mean_daily_returns = daily_returns.mean()
```

```python
# Calculate the covariance matrix of daily returns
cov_matrix = daily_returns.cov()
```

```python
# Display the results
print("📈 Mean Daily Returns:")
mean_daily_returns
```

```python
print("📊 Covariance Matrix of Returns:")
cov_matrix
```

```python
# Set seed for reproducibility
np.random.seed(42)
```

```python
# Simulation setup
simulated_portfolios = []  # To store final cumulative return of each simulation
```

```python
# Run Monte Carlo simulations
for _ in range(num_simulations):
    # Generate simulated daily returns for each stock (shape: 252 days × 4 stocks)
    simulated_daily_returns = np.random.multivariate_normal(mean_daily_returns, cov_matrix, trading_days)
    # Compute portfolio return per day by multiplying each row by the weights
    portfolio_daily_returns = simulated_daily_returns.dot(weights)  # Shape: (252,)
    # Compute cumulative portfolio return over the year
    cumulative_return = np.cumprod(1 + portfolio_daily_returns)[-1] - 1  # Final return over 1 year
    # Append result
    simulated_portfolios.append(cumulative_return)
```

```python
# Convert to numpy array for further analysis
simulated_portfolios = np.array(simulated_portfolios)
```

```python
# Display summary stats
print(f"Expected Final Portfolio Return (mean): {simulated_portfolios.mean():.4f}")
print(f"Standard Deviation of Returns: {simulated_portfolios.std():.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 6: Analyze simulation results
mean_sim_return = simulated_portfolios.mean()
std_sim_return = simulated_portfolios.std()
```

```python
# Step 7: Calculate VaR and CVaR at 95% confidence level
var_95 = np.percentile(simulated_portfolios, 5)
cvar_95 = simulated_portfolios[simulated_portfolios <= var_95].mean()
```

```python
# Step 8: Print all four results
print(f"📈 Mean Final Portfolio Return: {mean_sim_return:.4f}")
print(f"📉 Standard Deviation of Returns: {std_sim_return:.4f}")
print(f"🔻 Value at Risk (95% Confidence): {var_95:.4f}")
print(f"⚠️ Conditional Value at Risk (CVaR): {cvar_95:.4f}")
```

```python
# Step 9: Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(simulated_portfolios, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
# Highlight VaR and Mean
plt.axvline(var_95, color='red', linestyle='--', linewidth=2, label='VaR 95%')
plt.axvline(mean_sim_return, color='green', linestyle='--', linewidth=2, label='Mean Return')
# Add labels and legend
plt.title("Monte Carlo Simulation - Portfolio Return Distribution")
plt.xlabel("Final Portfolio Return")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```
